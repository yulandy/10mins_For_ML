```python
# Prompt: 產生使用gradient descent的Python範例，將搜尋過程繪圖: x代表參數theta，y代表 cost ，標示每次疊代過程中的parameters

import numpy as np
import matplotlib.pyplot as plt

# 定義目標函數
def cost_function(theta):
    return theta**2 + 3*theta + 2

# 定義目標函數的導數
def gradient(theta):
    return 2*theta + 3

# 梯度下降算法
def gradient_descent(starting_theta, learning_rate, iterations):
    theta = starting_theta
    theta_history = []
    cost_history = []

    for _ in range(iterations):
        theta_history.append(theta)
        cost_history.append(cost_function(theta))

        theta = theta - learning_rate * gradient(theta)

    return theta_history, cost_history

# 初始化參數
starting_theta = 8  # 初始值
learning_rate = 0.2  # 學習率
iterations = 9  # 迭代次數

# 執行梯度下降
theta_history, cost_history = gradient_descent(starting_theta, learning_rate, iterations)

# 繪製結果
theta_values = np.linspace(-4, starting_theta, 400)
cost_values = cost_function(theta_values)

plt.figure(figsize=(7, 5))
plt.plot(theta_values, cost_values, label='Cost Function')
plt.scatter(theta_history, cost_history, color='red', label='Iterations', zorder=5)
for i, (theta, cost) in enumerate(zip(theta_history, cost_history)):
    plt.annotate(f'{i}', (theta, cost), textcoords="offset points", xytext=(0,5), ha='center')

plt.title('Gradient Descent')
plt.xlabel(r'Parameter $\theta$')
plt.ylabel(r'Cost Function Value $f(\theta)$')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

# 定义梯度下降函数
def gradient_descent(f, df, theta_init, learning_rate, num_iterations):
    theta = theta_init
    theta_history = [theta_init]

    for iteration in range(num_iterations):
        gradient = df(theta)
        theta = theta - learning_rate * gradient
        theta_history.append(theta)

    return np.array(theta_history)

# 定义函数及其导数
def f1(theta):
    return theta[0]**2 + theta[1]**2

def df1(theta):
    return 2 * theta

def f2(theta):
    return 10 * theta[0]**2 + theta[1]**2

def df2(theta):
    return np.array([20 * theta[0], 2 * theta[1]])

# 初始参数、学习率和迭代次数
theta_init = np.array([3.0, 4.0])
learning_rate = 0.01
num_iterations = 75
# 运行梯度下降
theta_history1 = gradient_descent(f1, df1, theta_init, learning_rate, num_iterations)
theta_history2 = gradient_descent(f2, df2, theta_init, learning_rate, num_iterations)

# 绘制等高线和参数迭代过程
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# 绘制第一个函数的等高线
theta1 = np.linspace(-5, 5, 100)
theta2 = np.linspace(-5, 5, 100)
Theta1, Theta2 = np.meshgrid(theta1, theta2)
Y1 = Theta1**2 + Theta2**2
contour1 = ax1.contour(Theta1, Theta2, Y1, levels=np.linspace(np.min(Y1), np.max(Y1), 50), cmap='viridis')
ax1.set_xlabel(r'$\theta_1$')
ax1.set_ylabel(r'$\theta_2$')
ax1.set_title(r'Contours of $J = \theta_1^2 + \theta_2^2$')

# 绘制第一个函数的参数迭代路径
ax1.plot(theta_history1[:, 0], theta_history1[:, 1], marker='o', color='r', label='Gradient Descent')
ax1.legend()

# 绘制第二个函数的等高线
Y2 = 10 * Theta1**2 + Theta2**2
contour2 = ax2.contour(Theta1, Theta2, Y2, levels=np.linspace(np.min(Y2), np.max(Y2), 50), cmap='viridis')
ax2.set_xlabel(r'$\theta_1$')
ax2.set_ylabel(r'$\theta_2$')
ax2.set_title(r'Contours of $J = 10\theta_1^2 + \theta_2^2$')

# 绘制第二个函数的参数迭代路径
ax2.plot(theta_history2[:, 0], theta_history2[:, 1], marker='o', color='r', label='Gradient Descent')
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# 生成数据
np.random.seed(0)
X1 = np.random.normal(1, 2, 100)
X2 = np.random.normal(100, 5, 100)
X = np.c_[X1, X2]
y = 100 * X[:, 0] + X[:, 1] + np.random.normal(1, 2, 100)


# 特征缩放
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_b = np.c_[np.ones((100, 1)), X_scaled]



# 定义梯度下降函数
def gradient_descent(X_b, y, learning_rate, num_iterations):
    m = len(y)
    theta = np.random.randn(3)  # 随机初始化参数
    theta_history = []
    mse_history = []

    for iteration in range(num_iterations):
        gradients = 2 / m * X_b.T.dot(X_b.dot(theta) - y)
        theta = theta - learning_rate * gradients

        # 记录当前参数
        theta_history.append(theta)

        # 计算当前参数下的均方误差（MSE）
        y_pred = X_b.dot(theta)
        mse = np.mean((y_pred - y) ** 2)
        mse_history.append(mse)

    return theta, mse_history, theta_history

# 初始学习率和迭代次数
learning_rate = 0.1     # 特徵縮放後，可以使用較大的學習率，加快收斂性
num_iterations = 20

# 运行梯度下降
theta_best, mse_history, theta_history = gradient_descent(X_b, y, learning_rate, num_iterations)



# 绘制 MSE 随迭代次数的变化图像
plt.figure(figsize=(10, 6))
plt.plot(np.arange(1, num_iterations + 1), mse_history, color='b', marker='o', linestyle='-', linewidth=2)
plt.xlabel('Iterations')
plt.ylabel('Mean Square Error (MSE)')
plt.title(f'Feature Sclaing; Learning Rate = {learning_rate}')
plt.grid(True)
plt.show()



```
