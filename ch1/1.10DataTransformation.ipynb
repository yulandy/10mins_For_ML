```python
import numpy as np
import matplotlib.pyplot as plt

# Generate right-skewed data
np.random.seed(0)
data = np.random.gamma(shape=2, scale=1, size=1000)

# Plot the histogram of original data
plt.figure(figsize=(15, 5))

# Original Data
plt.subplot(1, 3, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Mild Right-Skewed Data')
plt.xlabel('$x$')
plt.ylabel('Frequency')

# Apply square root transformation
sqrt_data = np.sqrt(data)

# Plot the histogram of square root-transformed data
plt.subplot(1, 3, 2)
plt.hist(sqrt_data, bins=30, color='salmon', edgecolor='black', alpha=0.7)
plt.title('Square Root-Transformed Data')
plt.xlabel('$\sqrt{x}$')
plt.ylabel('Frequency')

# Apply log transformation
log_data = np.log(data)

# Plot the histogram of log-transformed data
plt.subplot(1, 3, 3)
plt.hist(log_data, bins=30, color='green', edgecolor='black', alpha=0.7)
plt.title('Log-Transformed Data')
plt.xlabel('$\log x$')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate log-normal distributed data
np.random.seed(0)
log_normal_data = np.random.lognormal(mean=0, sigma=1, size=1000)

# Plot the histogram of original log-normal distributed data
plt.figure(figsize=(15, 5))

# Original Data
plt.subplot(1, 3, 1)
plt.hist(log_normal_data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Moderate Right-Skewed Data Distribution')
plt.xlabel('$x$')
plt.ylabel('Frequency')


# Apply square root transformation
sqrt_log_normal_data = np.sqrt(log_normal_data)

# Plot the histogram of square root-transformed data
plt.subplot(1, 3, 2)
plt.hist(sqrt_log_normal_data, bins=30, color='green', edgecolor='black', alpha=0.7)
plt.title('Square Root-Transformed Data Distribution')
plt.xlabel('$\sqrt{x}$')
plt.ylabel('Frequency')


# Apply log transformation
log_log_normal_data = np.log(log_normal_data)

# Plot the histogram of log-transformed data
plt.subplot(1, 3, 3)
plt.hist(log_log_normal_data, bins=30, color='salmon', edgecolor='black', alpha=0.7)
plt.title('Log-Transformed Data Distribution')
plt.xlabel('$\log x$')
plt.ylabel('Frequency')



plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skewnorm

# 1. Generate left-skewed data
# Define the parameters for skewness
a = -5  # Negative value for left skew
size = 1000

# Generate the skewed data using scipy's skewnorm
x = skewnorm.rvs(a, size=size)


plt.figure(figsize=(15, 5))

# Plot the skewed data
plt.subplot(1, 3, 1)
plt.hist(x, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Left-Skewed Data')
plt.xlabel('x')
plt.ylabel('Frequency')

# 2. Data Reflection
max_x = np.max(x)
reflected_x = (max_x + 1) - x

# Plot the reflected data
plt.subplot(1, 3, 2)
plt.hist(reflected_x, bins=30, color='green', edgecolor='black', alpha=0.7)
plt.title('Reflected Data')
plt.xlabel('(max+1)-x')
plt.ylabel('Frequency')

# 3. Log Transformation
# Adding a constant before applying log transformation to handle potential 0 values
log_transformed_x = np.log(reflected_x + 1)

# Plot the log-transformed data
plt.subplot(1, 3, 3)
plt.hist(log_transformed_x, bins=30, color='salmon', edgecolor='black', alpha=0.7)
plt.title('Log-Transformed Data')
plt.xlabel('log((max+1)-x + 1)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(0.5, 100, 1000)  # Generate 100 data points from 1 to 10

# Compute log(x) and sqrt(x)
log_x = np.log(x)
sqrt_x = np.sqrt(x)

# Plotting
plt.figure(figsize=(8, 6))

# Plot log(x)
plt.plot(x, log_x, label='$\log x$', color='skyblue', linewidth=2)

# Plot sqrt(x)
plt.plot(x, sqrt_x, label='$\sqrt{x}$', color='salmon', linewidth=2)

# Add labels and title
plt.xlabel('x')
plt.ylabel('Transformed Values')
plt.title('Logarithm vs Square Root Transformation')
plt.legend()

plt.grid(True)
plt.show()

```
