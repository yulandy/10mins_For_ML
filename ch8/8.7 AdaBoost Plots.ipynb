```python
import numpy as np
import matplotlib.pyplot as plt

# Define r values avoiding r=1 to prevent division by zero and undefined log values
r = np.linspace(0.01, 0.99, 500)  # Slightly away from 0 and 1 to avoid undefined points

# Define the function
y = np.log( (1 -r)/r)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(r, y, label=r'$\alpha=\log(\frac{1- \epsilon }{\epsilon})$', color='b')

# Add labels and title

plt.xlabel(r'$\epsilon$', fontsize=12)
plt.ylabel(r'$\alpha$', fontsize=12)

# Show grid and legend
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```

```python
# prompt: plot exp(\alpha)  where \alpha in [-5,5]

import numpy as np
import matplotlib.pyplot as plt

# Define alpha values
alpha = np.linspace(-4, 4, 100)

# Calculate exp(alpha)
y = np.exp(alpha)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(alpha, y, label=r'$e^{\alpha}$', color='b')

# Add labels and title
plt.xlabel(r'$\alpha$', fontsize=12)
plt.ylabel(r'$e^{\alpha}$', fontsize=12)


# Show grid and legend
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```
