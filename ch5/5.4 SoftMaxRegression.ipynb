```python
# sklearn 加載iris資料庫，花瓣長度和花瓣寬度為特徵，用softmax regression分類和圖示化分類結果；三個legends，個別標示三個類別 Iris verginica, Iris versicolor, Iris setosa
# 程式微調整: 不將資料分成訓練和測試集
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# 加载 Iris 数据集
iris = load_iris()
X = iris.data[:, 2:4]  # 选择花瓣长度和花瓣宽度作为特征
y = iris.target

# 定义 Softmax 回归模型
# softmax regression 也稱 multinomial logistic regression；C值用來控制模型的複雜度，避免過度或欠擬合
softmax_reg = LogisticRegression(multi_class="multinomial", solver="lbfgs", C=10, random_state=42)

# 训练模型
softmax_reg.fit(X, y)

# 在訓練集上进行预测
y_pred = softmax_reg.predict(X)
# 计算准确率
accuracy = accuracy_score(y, y_pred)
print("準確率:", accuracy)

# 可视化分类结果
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
Z = softmax_reg.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.4)  # 根據預測類別0、1、2，繪製等高線

# 绘制每个类别的数据点，并为每个类别添加图例标签
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], label='Iris setosa', marker='o')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], label='Iris versicolor', marker='^')
plt.scatter(X[y == 2][:, 0], X[y == 2][:, 1], label='Iris virginica', marker='s')

plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.title('Softmax Regression')
plt.legend(loc='best')
plt.show()
```
