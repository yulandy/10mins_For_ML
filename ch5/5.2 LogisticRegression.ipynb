```python
#  python sklearn 加載iris資料庫，並圖示化前兩個特徵和後兩個特徵，並加legend標示setosa, versicolor, or virginica
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# 加载iris数据集
iris = load_iris()
X = iris.data[:, [0, 1, 2, 3]]  # 选择所有样本的前两个特征和后两个特征
y = iris.target
target_names = iris.target_names

# 绘制散点图
plt.figure(figsize=(12, 5))

# 绘制前两个特征的散点图
plt.subplot(1, 2, 1)
for i, target_name in enumerate(target_names):
    plt.scatter(X[y == i, 0], X[y == i, 1], label=target_name)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title('First two features')
plt.legend()

# 绘制后两个特征的散点图
plt.subplot(1, 2, 2)
for i, target_name in enumerate(target_names):
    plt.scatter(X[y == i, 2], X[y == i, 3], label=target_name)
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.title('Last two features')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 花瓣寬度(petal width)作為特徵,用Logistic Regression將Iris Setosa (class 0) 與其他品種進行二元分類,並在同一張圖上繪製樣本點、決策邊界曲線,以及機率為0.5的直線
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris

# 載入Iris數據集
iris = load_iris()
X = iris.data[:, 3]  # 花瓣寬度
y = (iris.target != 0).astype(int)  # Iris Setosa為0,其他為1

# 建立Logistic Regression模型
log_reg = LogisticRegression()
log_reg.fit(X.reshape(-1, 1), y)   # Reshape your data either using array.reshape(-1, 1) if your data has a single feature

# 繪製樣本點 (很多資料有相同的petal width)
plt.figure(figsize=(8, 6))
plt.scatter(X[y == 0], np.zeros_like(X[y == 0]), label="Iris Setosa")
plt.scatter(X[y == 1], np.ones_like(X[y == 1]), label="Others")

# 繪製決策邊界
X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_prob = log_reg.predict_proba(X_new)
plt.plot(X_new.ravel(), y_prob[:, 1], "g--", label="Iris Setosa boundary")

# 繪製機率為0.5的直線
threshold = -log_reg.intercept_[0] / log_reg.coef_[0][0]
plt.axvline(threshold, color='k', linestyle='--', label='Probability = 0.5')


# 設定圖例和標題
plt.legend()
plt.xlabel("Petal Width")
plt.ylabel("Target Class")
plt.title("Iris Setosa Classification Using Petal Width")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris

# 載入Iris數據集
iris = load_iris()
X = iris.data[:, 2:4]  # 花瓣長度和花瓣寬度
y = (iris.target == 2).astype(int)  # Iris Virginica為1,其他為0

# 建立Logistic Regression模型
log_reg = LogisticRegression()
log_reg.fit(X, y)

# 繪製樣本點
plt.figure(figsize=(8, 6))
plt.scatter(X[y == 0, 0], X[y == 0, 1], label="Others")
plt.scatter(X[y == 1, 0], X[y == 1, 1], label="Iris Virginica")

# 繪製決策邊界
x1_plot = np.linspace(0, 8, 1000)
x2_plot = np.linspace(0, 3, 1000)
X1, X2 = np.meshgrid(x1_plot, x2_plot)
X_plot = np.column_stack((X1.ravel(), X2.ravel()))
y_prob = log_reg.predict_proba(X_plot)
y_prob = y_prob[:, 1].reshape(X1.shape)
plt.contour(X1, X2, y_prob, levels=[0.5], colors='k', linestyles='--', label='Probability = 0.5')

# 設定圖例和標題
plt.legend()
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Iris Virginica Classification Using Petal Features")
plt.show()
```

```python
# petal length, petal width, sepal width作為特徵,用Logistic Regression將Iris Versicolor與其他品種進行二元分類,
# 在同一張圖上繪製樣本點,以及inner product of theta (parameter vector) and x (feature vector) equals 0 的平面

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris

# 載入Iris數據集
iris = load_iris()
X = iris.data[:, [2, 3, 1]]  # 花瓣長度、花瓣寬度和花萼寬度
y = (iris.target == 1).astype(int)  # Iris Versicolor為1,其他為0

# 建立Logistic Regression模型
log_reg = LogisticRegression()
log_reg.fit(X, y)

# 繪製樣本點
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[y == 0, 0], X[y == 0, 1], X[y == 0, 2], label="Others")
ax.scatter(X[y == 1, 0], X[y == 1, 1], X[y == 1, 2], label="Iris Versicolor")

# 繪製theta.T @ x = 0的平面
theta = log_reg.coef_[0]
intercept = log_reg.intercept_[0]
x1_mesh, x2_mesh = np.meshgrid(np.linspace(0, 8, 100), np.linspace(0, 3, 100))
x3_mesh = (-theta[0] * x1_mesh - theta[1] * x2_mesh - intercept) / theta[2]
ax.plot_surface(x1_mesh, x2_mesh, x3_mesh, alpha=0.3, color='r', label='theta.T @ x = 0')

# 設定圖例和標題
ax.set_xlabel("Petal Length")
ax.set_ylabel("Petal Width")
ax.set_zlabel("Sepal Width")
ax.set_title("Iris Versicolor Classification Using Petal and Sepal Features")
plt.show()
```
