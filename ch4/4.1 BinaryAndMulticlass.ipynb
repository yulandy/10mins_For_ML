```python
# 超平面繪圖
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 生成示例数据
np.random.seed(0)
num_samples = 100
X = np.random.rand(num_samples, 3)  # 三维特征
y = np.where(X[:, 0] + X[:, 1] - X[:, 2] > 0, 1, 0)  # 使用一个平面将数据分成两类

# 分类为0和1的数据
X_class0 = X[y == 0]
X_class1 = X[y == 1]

# 绘制图形
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 绘制数据点
ax.scatter(X_class0[:, 0], X_class0[:, 1], X_class0[:, 2], c='b', marker='o', label='Class 0')
ax.scatter(X_class1[:, 0], X_class1[:, 1], X_class1[:, 2], c='r', marker='^', label='Class 1')

# 绘制分割平面
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
X_plane, Y_plane = np.meshgrid(x, y)
Z_plane = X_plane + Y_plane  # 这是一个简单的平面方程，可根据需要进行调整
ax.plot_surface(X_plane, Y_plane, Z_plane, alpha=0.5, color='gray')

# 设置坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# 设置图例
ax.legend()

plt.title('3D Classification Problem')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成三类数据
num_samples = 100
class_1_mean = [1, 1]
class_2_mean = [3, 3]
class_3_mean = [5, 1]
class_1 = np.random.randn(num_samples, 2) + class_1_mean
class_2 = np.random.randn(num_samples, 2) + class_2_mean
class_3 = np.random.randn(num_samples, 2) + class_3_mean

# 绘制散点图
plt.scatter(class_1[:, 0], class_1[:, 1], marker='o', label='Class 1')
plt.scatter(class_2[:, 0], class_2[:, 1], marker='x', label='Class 2')
plt.scatter(class_3[:, 0], class_3[:, 1], marker='^', label='Class 3')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Three Classes in 2D Space')
plt.legend()
plt.grid(True)
plt.show()
```
